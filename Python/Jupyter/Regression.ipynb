<div stlye="font-size: 1.5rem; margin:15px 0;">
    <strong>A return to a former or less devloped state</strong><br/>
    basics: in 2 dimensions, best fit line<br/><br/>
    <strong>Simple Linear Regression</strong><br/>
    Independent: Predictor | Dependent: Output<br/><br/>
    Goal is to minimize prediction error via the following: 
</div>

${\displaystyle{
    \text{-----------------------}\\
    \text{Error Minimization: } \\
    \text{----------------------} \\
    Err = \sum_{i=1}^N(y_i - a_0 - a_1x_i)^2 \\
    a_1 = \frac{\sum_ix_iy_i - (\sum_ix_i)(\sum_iy_i) / n}{\sum_i x_i^2 - (\sum_ix_i)^2 / n} \\
    a_0 = \bar{y} - a_1\bar{x} \\
    \text{-----------------------}\\
    \text{Sum of square error: } \\
    \text{-----------------------}\\
    SSE = \sum(y_i-\hat{y_i})^2 \\
    \text{-----------------------}\\
    \text{SSR = sum of squares regression, SST = sum of squares total: } \\
    \text{-----------------------}\\
    SSR = \sum(\hat{y_i} - \bar{y-i})^2 \\
    SST = \sum(y_i - \bar{y_i})^2 \\
    R^2 = \frac{SSR}{SST} \\
    \text{-----------------------}\\
}}$

<div style="margin: 20px 0;">
    Sum of square error AKA standard error of the estimate.<br/><br/>
    <strong>Homoskedasticity:</strong> When residuals show more or less constant variance<br/>
    <strong>Heteroskedasticity:</strong> When residuals variance is divergent<br/><br/>
    <strong>Non-Linear Regression</strong><br/>
    Below are a few simplistic regression models: 
</div>

${\displaystyle{
    \text{-----------------------}\\
    \text{Quadratic, Logrithmic, and Exponential models (Q, L, E)} \\
    \text{-----------------------}\\
    Q = b_0 + b_1x + b_2x^2 \\
    L = a + b (\ln x)\\
    E = a^*b^x \\
    \text{-----------------------}\\
}}$

<div style="margin: 20px 0;">
    <strong>Multiple Linear Regression</strong><br/>
    Used when number of predictor variables exceeds one<br/>
    Equations used below: <br/>
</div>

${\displaystyle{
    \text{-----------------------}\\
    \text{Multiple Regression of y on } x_1, x_2, \ldots, x_k \\
    \text{please note: } a_0 \text{ stands for a constant bias you choose} \\
    \text{-----------------------}\\
    y = a_0 + a_1x_1 + a_2x_2 + \ldots + a_kx_k \\
    \text{-----------------------}\\
    \text{Find the minimization of a by: }\\
    \text{-----------------------}\\
    j(a) = ||Y - Xa||^2 \\
    \text{-----------------------}\\
    \text{example problem: creating the 'Psuedo-inverse'}\\
    \text{-----------------------}\\
    y = xa \rightarrow ||y - xa||^2 \\
    2x^t(y - xa) = 0 \rightarrow x^ty = x^txa \\
    a = (x^tx)^{-1}x^ty = x^*y \\
    x^* = (x^tx)^{-1}x^t \\
    \text{-----------------------}\\
    \text{Linear Regression functions: Model, Parameters and Cost (J = cost)}\\
    \text{-----------------------}\\
    \hat{y} = w_0 + w^tx \text{ with params } w_0, w \\
    j(w_0, w) = \frac{1}{2} \sum_{i=1}^n(y_i - (w_0 + w^tx_i))^2 \\
    \frac{\partial J}{\partial w_0} = -\sum_{i=1}^n \text{to be cont.}\\
}}$

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

N = 32
alpha = 1.3
beta = np.array([0.5, 1.9])
x0 = np.ones((N, 1))
x_data = np.random.randn(N, 2)
X = np.hstack((x0, x_data))
Y = x_data.dot(beta) + alpha + 0.1 * np.random.randn(1)
w = np.random.randn(3)

# drop off highly dependent on learning rate, which is chosen at random
# increasing past the optimal rate increases error loss
learning_rate = 1e-3
result = []
loss = 0
for t in range(200):
    y_pred = X.dot(w)
    loss = np.square(y_pred - Y).sum()
    if t % 10 == 0:
        print(t, loss)
        result.append(loss)
    grad_y_pred = 2.0 * (y_pred - Y)
    grad_w = X.T.dot(grad_y_pred)
    
    w -= learning_rate * grad_w
print(w)
plt.plot(np.arange(0,200,10), result)
plt.show()
```

<div stlye="font-size: 1.5rem; margin:15px 0;"> 
    Logistic regression used for classification, linear used for numeric<br/>
    Using the following: 
</div>

${\displaystyle{
    \text{----------------------------------------------------} \\
    \text{Log odds: } \\
    \text{----------------------------------------------------} \\
    p = \frac{e^{(a_0+a_1x_1+ \ldots +a_kx_k)}}{1 + e^{(a_0+a_1x_1+ \ldots +a_kx_k)}} \\
    log(\frac{p}{1-p}) = a_0+a_1x_1+ \ldots +a_kx_k \\
    \text{----------------------------------------------------} \\
}}$
