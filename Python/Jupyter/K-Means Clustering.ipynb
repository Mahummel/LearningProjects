    <div style="font-weight: bold; font-size: 20px; margin-bottom: 10px;">Section 1: K-Means Objective</div> 
    <div style="margin-left: 15px;">
        word2vec -- natural language processing <br/>
        Autoencoder -- image processing <br/>
        Clustering -- assign each data point to one cluster: [z<sub>1</sub>, ... , z<sub>n</sub>] where z<sub>i</sub> exists in 
        {1,2, ... , k} <br/>
        Inter-cluster distance -- distance between points in clusters 1 and 2 <br/>
        Intra-clluster distance -- distance between points in singular cluster <br/><br/>
        Goal is to maximize inter-cluster distance, and to minimize intra-cluster distance <br/><br/>   
$$
\begin{align}
||\phi(x_{i}) - \mu_{zi}||\\
\phi = feature\:extractor\\
\phi(x_{i})\; is\: a\: feature\: vector\\
\end{align}
$$
        <br/><br/>
        Take z<sub>i</sub> = x<sub>i</sub> - distance to mu<sub>i</sub>, this helps classify which cluster it belongs to.<br/>
        ex:<br/><br/>
$$
\begin{align}
X_i = 0,\:2,\:5,\:7,\:12,\:14 \\
Let: \; \mu_{1} = 1,\; \mu_{2} = 6,\; \mu_{3} = 13\\ \\
X_1: Z_1 &= ||X_1 - \mu_1|| = || 0 - 1 || = 1 \\
&= ||X_1 - \mu_2|| = || 0 - 6 || = 6\\
&= ||X_1 - \mu_3|| = || 0 - 13 || = 13\\
Z_1 &= 1 \rightarrow \mu_1 \\ \\
X_2: Z_2 &= ||X_2 - \mu_1|| = || 2 - 1 || = 1 \\
&= ||X_2 - \mu_2|| = || 2 - 6 || = 4\\
&= ||X_2 - \mu_3|| = || 2 - 13 || = 11\\
Z_2 &= 1 \rightarrow \mu_1 \\ \\
X_3: Z_3 &= ||X_3 - \mu_1|| = || 5 - 1 || = 4 \\
&= ||X_3 - \mu_2|| = || 5 - 6 || = 1\\
&= ||X_3 - \mu_3|| = || 5 - 13 || = 8\\
Z_3 &= 1 \rightarrow \mu_2 \\ \\
Z\:number\:refers\:to\:cluster \\
Used\:if\:we\:know\:centroids(clusters)
\end{align}
$$
        <br/><br/>
        If we don't know centroid values, but know the assignments, we can find the centroids via the following:<br/><br/>
$$
\begin{align}
Let: \: z_1=z_2=1\\
z_3=z_4=2\\ 
z_5=z_6=3\\
X_i\: above \\ \\
\end{align}
$$
$$
\begin{align}
\mu_1 &= \mu^2 + (2-\mu)^2 \\
&= \mu^2 + \mu^2 - 4\mu + 4 \\
&= 2\mu^2 - 4\mu +4 \\
&= 2(\mu^2 - 2\mu + 2) \\
&= 2[(\mu - 1)^2 + 1] \\
&= 1
\end{align}
$$
        General equation for above is the following:<br/><br/>
$$
\mu_i = argmin_\mu (X_1 - \mu)^2 + \ldots + (X_n - \mu)^2 \\
\forall X \in \mu_i \\ \\
$$
        What if we know neither?<br/><br/>
$$
\text{Given Centroids}\; \mu_1, \ldots, \mu_K, \;\text{assign each point to best centroid} \\
z_i \leftarrow arg min_{k = 1, \ldots, K} \; || \phi(x_i) - \mu_k ||^2 \\
\text{Given Cluster assignments} \; z_1, \ldots, z_n, \; \text{find the best centroids} \; \mu_1,\ldots,\mu_k \\
\mu_k \leftarrow \frac{1}{|\{i:z_i = k\}|} \sum_{i:z_i=k} \phi(x_i) \\
$$
